```python
import matplotlib.pyplot as plt
import os


iterations = [10, 20, 50, 100, 200]
win_rates_black = []
win_rates_white = []

for i in iterations:
    with open(f'results/MCTS_black_{i}_iterations.txt', 'r') as f:
        results = f.read().splitlines()
        # Get the last line
        results = results[-1]
        start = results.find(': ') + 2
        end = results.find('%')
        win_rates_black.append(float(results[start:end]))


    with open(f'results/MCTS_white_{i}_iterations.txt', 'r') as f:
        results = f.read().splitlines()
        # Get the last line
        results = results[-1]
        start = results.find(': ') + 2
        end = results.find('%')
        win_rates_white.append(100 - float(results[start:end]))

# ggplot style
plt.style.use('ggplot')
# plot with two legends (one for each line)
plt.plot(iterations, win_rates_black, label='Play as Black')
plt.plot(iterations, win_rates_white, label='Play as White')
plt.legend()
plt.xlabel('Iterations')
plt.ylabel('Win rate (%)')
plt.title('MCTS Win rate against DQN')
plt.savefig('results/win_rate_vs_iterations.png')
plt.show()
```
